### Step 11. What was the most ordered item in the choice_description column?

```python
unique_elements = set(num_of_choice_description)
frequency_dict = {}
most_repeated_element = None
highest_frequency = 0
second_most_repeated_element = None
second_highest_frequency = 0

for element in unique_elements:
    frequency = num_of_choice_description.count(element)
    frequency_dict[element] = frequency

    if frequency > highest_frequency:
        second_most_repeated_element = most_repeated_element
        second_highest_frequency = highest_frequency
        most_repeated_element = element
        highest_frequency = frequency
    elif frequency > second_highest_frequency and element != most_repeated_element:
        second_most_repeated_element = element
        second_highest_frequency = frequency

# print("Frequency of each element:")
# for element, frequency in frequency_dict.items():
#     print(element, ":", frequency)

# print("Most repeated element:", most_repeated_element)
# print("Frequency of the most repeated element:", highest_frequency)

print("the most ordered item in the choice_description column:", second_most_repeated_element)
print("Frequency of most repeated element:", second_highest_frequency)
```

### Step 13. Turn the item price into a float

```python
chickisthereNULLS = set(chipo['item_price'].isna())
chickisthereNULLS
```

```python
type(chipo['item_price'])
```

```python

def convert_to_float(string_value):
    cleaned_string = string_value.replace('$', '')
    return float(cleaned_string)

# Apply the conversion function to the Series
chipo['item_price'] = chipo['item_price'].apply(convert_to_float)

# Print the float Series
print(chipo['item_price'])

```

### Step 16. What is the average revenue amount per order?

```python
# Solution 1
###"Average revenue" refers to the total amount of money earned divided by the number of orders.

avg_revenue = chipo['revenue'].sum() / len(set(chipo['order_id']))
avg_revenue
```

```python
# Solution 2
average_revenue_per_order = chipo.groupby('order_id')['revenue'].mean()

average_revenue_per_order
```
